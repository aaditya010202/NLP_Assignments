#USING TENSORFLOW



##Part A



> Seperating the training and testing data



```python
import tensorflow as tf
from tensorflow import keras
```

```python
fashion_mnist=keras.datasets.fashion_mnist
(X_train,y_train),(X_test,y_test)=fashion_mnist.load_data()
```



> making a list for the class names



```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```



> Extracting the data with label 0(T-shirts/tops) and 1(trousers) only



```python
X_new_train=[]
y_new_train=[]
X_new_test=[]
y_new_test=[]

for i in range(60000):
  if y_train[i]==0 or y_train[i]==1:
    X_new_train.append(X_train[i]/255) #to normalise the values
    y_new_train.append(y_train[i])

for i in range(10000):
  if y_test[i]==0 or y_test[i]==1:
    X_new_test.append(X_test[i]/255)
    y_new_test.append(y_test[i])

```

```python
import numpy as np
X_new_train=np.array(X_new_train)
X_new_test=np.array(X_new_test)
y_new_train=np.array(y_new_train)
y_new_test=np.array(y_new_test)
```

```python
#TRAINING IMAGES
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))

for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(X_new_train[i], cmap=plt.cm.binary)
  plt.xlabel(class_names[y_new_train[i]])
plt.show()
```

```python
#TESTING IMAGES
plt.figure(figsize=(10,10))

for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(X_new_test[i], cmap=plt.cm.binary)
  plt.xlabel(class_names[y_new_test[i]])
plt.show()
```

##Part B






```python
simple_model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(1,activation="sigmoid"),
])
```

```python
simple_model.compile(optimizer="adam",
                     loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                     metrics=["accuracy"])
```

```python
simple_model.fit(X_new_train,y_new_train,epochs=10)
```

```python
test_loss,test_acc=simple_model.evaluate(X_new_test,y_new_test)
```

##Part C

```python
simple_model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation="relu"),
    keras.layers.Dense(100,activation="relu"),
    keras.layers.Dense(1,activation="sigmoid")
])
```

```python
simple_model.compile(optimizer="adam",
                    #  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                     loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                     metrics=["accuracy"])
```

```python
simple_model.fit(X_new_train,y_new_train,epochs=10)
```

```python
test_loss,test_acc=simple_model.evaluate(X_new_test,y_new_test)
```



> The accuracy increased in case of 2 hidden layers with 100 neurons
