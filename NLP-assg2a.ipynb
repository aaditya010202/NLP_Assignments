```python
words=['can', 'could', 'may', 'might', 'must', 'will']
from nltk.probability import ConditionalFreqDist
cfd=ConditionalFreqDist(
    (categories,w)
for categories in brown.categories()
for fileids in brown.fileids(categories)
for w in brown.words(fileids) if w in words

)
cfd
```

```python
cfd.plot()
```

```python
def frequent_words(text):
  fdist=FreqDist(text)
  return fdist.most_common(50)

frequent_words(brown.words(categories="news"))
```
