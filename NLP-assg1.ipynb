```python
from nltk.collocations import *
```

```python
#Bigrams
bigram=nltk.collocations.BigramAssocMeasures()
```

```python
finder_bigram=BigramCollocationFinder.from_words(text5)
finder_bigram.score_ngrams(bigram.likelihood_ratio)
```

```python
finder_bigram.nbest(bigram.likelihood_ratio,5)
```

```python
#Trigrams
#printing score
trigram_measures=nltk.collocations.TrigramAssocMeasures()
finder_trigram=TrigramCollocationFinder.from_words(text5)
finder_trigram.score_ngrams(trigram_measures.likelihood_ratio)
```

```python
#finding best trigrams
finder_trigram.nbest(trigram_measures.likelihood_ratio,5)
```
